Train all models and calculate the cosine similarities. This is done by first training the model and then running a function called calculate similarities and save this as a new column to the dataframe.

```python
#training a word2vec model with defautl parameters (sg = skipgram, vectorsize = 100, windowsize = 5)
model_default = Word2Vec(corpus_file=train_en, sg = 1 )
```

```python
def calculate_similarities(model, dataset ): 
    """function to calculate the cosine similarities through different word2vec models. 
    input variables: 
    model = the word 2vec model you trained
    dataset = the dataset you want to analyse (in our case the lexical_decision dataset)
    output: a list of all cosine similarities between the primes and pairs of your dataset, the amount of primes that are not in your vocabulary, the amount of target words that are not in your models vocabulary
    """
    vocabulary = list(model.wv.index_to_key)
    prime_count = 0
    target_count = 0
    cosine_similarities = []

    for i in range(0, dataset.shape[0]):
        if not dataset['prime'][i] in vocabulary: 
            prime_count +=1 
            cosine_similarities.append("prime_not_in_vocab")
            continue 
        if not dataset['target'][i] in vocabulary:
            target_count +=1 
            cosine_similarities.append("target_not_in_vocab")
            continue 
        cosine_similarities.append(model.wv.similarity(dataset['prime'][i], dataset['target'][i]))
    return cosine_similarities, prime_count, target_count



#cosine_sim, prim, targ = calculate_similarities(model_default, l_df)

    
```

```python
#find if there is a prime effect: 
#This is done by finding the index 

groups = l_df_nm.groupby(["prime", "isi"], sort = False)
helperlist = []
indexlist = []

for name, group in groups:
    #find the smallest reaction time
    smallest_RT_index = group['meanRT'].argmin()
    #find the index of the unrelated prime target pairs
    unrel_index = [ 1 if "unrel" in x else 0 for x in group['condition']]
    indexlist.append(group.index.values.flatten())
    #check if there is a priming effect or not 
    if not 0 in unrel_index:
        helperlist.append("no_comparison")
        continue
    if not 1 in unrel_index:
        helperlist.append("no_comparison")
        continue
    else:
        if smallest_RT_index == unrel_index.index(0):
            for i in range(0,len(unrel_index)):
                helperlist.append("prime_effect!")
        else: 
            for i in range(0,len(unrel_index)):
                helperlist.append("no_prime_effect")
    

#flatten the indexlist
indexlist2 = [indexlist[x].tolist() for x in range(0, len(indexlist)) ]
indexlist3 = list(np.concatenate(indexlist2).flat)
#sort the prime effect list by the index list to add to the original dataframe 
zipped_pairs = zip(indexlist3, helperlist)
  
z = [x for _, x in sorted(zipped_pairs)]
  
print(z)

l_df_nm['prime_effect'] = z
```
